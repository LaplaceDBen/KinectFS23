```python
simple_column = df['Simple Viewport Shading (ms)']
material_column = df['Material Viewport Shading (ms)']

max_simple = simple_column.max()
max_material = material_column.max()
min_simple = simple_column.min()
min_material = material_column.min()
mean_simple = simple_column.mean()
mean_material= material_column.mean()
median_simple = simple_column.median()
median_material = material_column.median()
std_simple = simple_column.std()
std_material = material_column.std()

print(f'Simple Viewport Shading:\nNumber of Inputs: 2000, Number of processed Inputs: {len(simple_column)}')
print(f'Maximum: {max_simple}ms, Minimum: {min_simple}ms, Mean: {mean_simple}ms, Median: {median_simple}ms, STD: {std_simple}ms')
print('--------------------------------------------------------------------------------------------------------------------------------')
print(f'Material Viewport Shading:\nNumber of Inputs: 2000, Number of processed Inputs: {len(material_column)}')
print(f'Maximum: {max_material}ms, Minimum: {min_material}ms, Mean: {mean_material}ms, Median: {median_material}ms, STD: {std_material}ms')
print('--------------------------------------------------------------------------------------------------------------------------------')
```
